# Load the data

Let's load the data and see what we are dealing with. Unless you are completely new to python or in data science, you know that we will use the pandas library to load our data inside the system. 

```python
import pandas as pd

training_csv_filename = "../files/train.csv"
# Training dataframe
df_tr = pd.read_csv(training_csv_filename)
df_tr.head(10)
```

```python
# Let's see the data type of each column
df_tr.dtypes
```

Pandas is pretty smart, and can automatically assume the types of certain columns, such as _RoomService_ -> Float or _Transported_ -> Boolean. 
The data types _object_ is used for any columns that don't match a starndard numerical data types (int, float or bool). Meaning that _string_ type are also considered as _object_ 

Now, you may have realised that the columns _CryoSleep_ and _VIP_ are considered as _object_. 
However, they seems to be _boolean_ type as well, like the _Transported_ column which has been cast as boolean. So, why the difference ? 
The difference comes in the NaN values (or non existing values). 

What are NaN values ? Several definition but it is basically when there is no values into a field. Let me explain

Let's see those na values for the field  _CryoSleep_ and _VIP_. Good thing pandas already come with some function to help us

```python
# Na value example 
df_tr[["CryoSleep","VIP"]][df_tr[["CryoSleep","VIP"]].isna()].head(10)
```

As we see, when there is na values, pandas fill the gap with "NaN" values which stands "Not a Number". 
Now let's took a look overall on how many NaN values exist for each column.

```python
# Search number of Na values per column
df_tr.isna().sum()
```

As we can see the fields _Transported_ does not have any NaN Values (normal since it is or y_true), and that is the reason why pandas is able to cast this column in boolean and not the columns _CryoSleep_ and _VIP_. Therefore, they are cast as _object_.

You might realise by now that the float fields (like _Age_) have NaN values but are cast as float. 

The reason is a technical, and I would not go in all the details but it is basically because float can handle NaN as being a float but bool cannot handle NaN values (it needs to be either True or False).
At least, it is like this for the current version of this current version of pandas, it might changes in the future. 

Now, you might wonder why I am talking about all of this. It is because Machine Learning algorithm can only work with number (one way or the other). 
However, in real life, data are not made of numbers (there is a lot of text or missing values like here) and the goal of the data preparation step is to transform those data in a state that a ML algoirthm can understand. 
One of early mistakes in ML algorithm is to not prepare your data correctly and your ML algorithm failed to train because it can't understand the data. That's why, whatever your dataset, you need to start by looking at the data field types and make sure the ML algo can understand it. For our case, it might be easy to believe that our columns _CryoSleep_ and _VIP_ are boolean and they will be easy to use, but because of the missing values, they are cast as string and if we do not understand the reason why they have been cast as such, we can prepare them correctly. 

In overall, you will notice that one of the best part of the job when creating a ML algorithms is to prepare the data (Filling the missing values, features enhancement , etc.). I would even say that a data set prepare correctly will have more impact on the prediction result than the ML algorithm you choose.
